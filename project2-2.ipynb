```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('carvan_train.csv')
```

```python
df.head()
```

```python
df['V86'].value_counts()
y=df['V86']
x=df.drop(['V86'],1)
```

```python
from sklearn.decomposition import KernelPCA
kpca=KernelPCA(n_components=30,kernel='rbf')
```

```python
k_x=kpca.fit_transform(x)
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE 

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_sample(k_x, y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X_res, y_res,test_size=0.33)
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train, y_train) 
```

```python
q=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, q)
```

```python
X_res.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
Dclf = DecisionTreeClassifier()
Dclf.fit(X_train, y_train) 
```

```python
q2=Dclf.predict(X_test)
accuracy_score(y_test, q2)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(Dclf, X_train, y_train))  
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,q2)
```

```python
from sklearn.metrics import fbeta_score
fbeta_score(y_test, q2, beta=2)
```
